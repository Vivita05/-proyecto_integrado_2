```python
# prompt: Armar grafico de barras con Promedio general de cada materia, ordenados de mayor a menor, agregar las faltas como valor constrastable en dos ejes. Agregar las leyendas que ayuden a la interpretación.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'Materia': ['Matemáticas', 'Física', 'Química', 'Historia', 'Literatura'],
    'Promedio': [8.5, 7.2, 9.1, 6.8, 8.0],
    'Faltas': [2, 5, 1, 7, 3]
}
df = pd.DataFrame(data)

# Sort the DataFrame by average grade in descending order
df_sorted = df.sort_values('Promedio', ascending=False)


# Create the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
bars = plt.bar(df_sorted['Materia'], df_sorted['Promedio'], color='skyblue', label='Promedio General')

# Add the number of absences as a contrasting value on a second y-axis
ax2 = plt.twinx()
ax2.plot(df_sorted['Materia'], df_sorted['Faltas'], marker='o', color='coral', linestyle='-', linewidth=2, label='Faltas')
ax2.set_ylabel('Número de Faltas', color='coral')
ax2.tick_params(axis='y', labelcolor='coral')

# Add labels and title
plt.xlabel('Materia')
plt.ylabel('Promedio General', color='skyblue')
plt.title('Promedio General por Materia y Número de Faltas')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Add value labels on top of each bar
for bar, value in zip(bars, df_sorted['Promedio']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{value:.1f}', ha='center', va='bottom')

# Add legend
plt.legend(loc='upper right')


# Add a grid for better visualization (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
